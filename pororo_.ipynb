# 3. 긍정 부정 퍼센트 계산

```python
############### 퍼센트 계산
def per(r):
  n=0 #부정
  p=0 #긍정 
  etc=0 #예외처리된 긴 문장

  for i in r:
    if i == "Negative":
      n+=1
    elif i == "Positive":
      p+=1
    else:
      etc+=1
  
  positiveP=(p/(len(r)-etc))*100
  negativeP=(n/(len(r)-etc))*100

  return positiveP,negativeP
```

```python
positive,negative = per(sample['result'])
```
